# Usage of matplotlib

## A small instance of bar graph

```python
import matplotlib.pyplot as plt
x1 = [2,4,6,8,10]
y1 = [8,6,2,5,6]
x2 = [1,3,5,7,9]
y2 = [5,2,7,8,2]
plt.bar(x1,y1, label='Bar1', color='c')
plt.bar(x2,y2, label='Bar1', color='r')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Instance of Mtaplotlib')
plt.legend()
plt.show()
```
